```python
# When passed a datetime column this functions splits out year, month, day

def get_date(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
```

```python
# RFM Modelling
# Importing necessary libraries
import squarify
```

```python
# We have to first create RFM model scores for each customer.
# Take the next date of our last trnasaction date to find the recency of each customer.
mostrecent_date = data_processed['InvoiceDate'].max() + dt.timedelta(days=1)
mostrecent_date
```

```python
# Find RFM Metrics for each customer
RFMScores = data_processed.groupby('CustomerID').agg({'InvoiceDate': lambda x: (mostrecent_date - x.max()).days, 'InvoiceNo': lambda x: len(x), 'Transaction Cost': lambda x: x.sum()})
RFMScores.head()
```

```python
# We have to now divide the RFM metrics into segments
# Lower the recency, higher the rating and higher the frequency and monetary value, higher the rating.

# Create labels for Recency, Frequency and Monetary
r_labels = range(4, 0, -1)
f_labels = range(1, 5)
m_labels = range(1,5)

# Assign these labels to 4 equal percentile groups 
r_groups = pd.qcut(RFMScores['Recency'], q=4, labels=r_labels) 
f_groups = pd.qcut(RFMScores['Frequency'], q=4, labels=f_labels)
m_groups = pd.qcut(RFMScores['Monetary'], q=4, labels=m_labels)

# Create new columns R, F and M 
RFMScores = RFMScores.assign(R = r_groups.values, F = f_groups.values, M = m_groups.values)
RFMScores.head()

# We can see that R, F and M columns have been created based on their values.
# Here 1 indicates worst performance and 4 indicates best performance.
# We can see thet Customer ID 12347 has been a valuable customer for the company. He has puchased just 2 days back, 
# which shows his high recency. Also he has made 182 trnasactions during his lifecycle and has given good monetary value.
# Therefore all his rankings are 4, 4 and 4.
```

```python
# Create a function to concatenate the 3 columns R, F and M
def concat_RFM(x):
    return str(x['R']) + str(x['F']) + str(x['M'])

# Create a new column of the concatenated values
RFMScores['RFM_Segment_Concat'] = RFMScores.apply(concat_RFM, axis=1)
RFMScores.head()

# We have created our RFM segments using their ranks in their respective columns.
```

```python
# We now want to see the number of RFM segments we created.
len(RFMScores.RFM_Segment_Concat.unique())

# We have created 61 RFM segments. But this does not make any sense to divide our customers into 61 segments.
# Instead we want to have 5 or 6 segments of our customers.
```

```python
# So our solution to the above problem is that we can now sum the R, F and M columns and then create segments of 
# our customers.
# Summing the 3 columns
RFMScores['RFM_Score'] = RFMScores[['R','F','M']].sum(axis=1)
RFMScores.head()

# Our next step is to use the RFM scores and define RFM levels for each score range.
```

```python
# Define rfm_level function
def rfm_level(df):
    if df['RFM_Score'] >= 10:
        return 'High Value Customers'
    elif df['RFM_Score'] >= 8:
        return 'Medium Value Customers'
    elif df['RFM_Score'] >= 5:
        return 'Promising Customers'
    else:
        return 'Require Activation'

# Create a new column RFM_Level
RFMScores['RFM_Level'] = RFMScores.apply(rfm_level, axis=1)
RFMScores.head()

# Each customer has been classified into one segment.
```

```python
# Calculate average values for each RFM_Level, and return the size of each segment 
rfm_level_agg = RFMScores.groupby('RFM_Level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(1)
print(rfm_level_agg)

# Findings from the segments

# We see that almost 50% (2127) of our customers lie in the top 2 tiers.So the company must be doing some right things so
# that 50% of customers are staying loyal with it. The company should therefore continue with the same existing strategies
# with the customers of the top 2 tiers.

# High Value Customers are those customers who have recently purchased and provide high monetary value. These group of 
# customers are of utmost important to the company. The company cannot afford to lose them. They must be retained by any
# means. Apt startegies should be formed so as to keep the customers happy and satisfied like free shipping and priority
# delivery.

# Medium Value Customers have the potential to become High Value Customers. These customers have shown interest in the
# company. Either they regularly visit the website or they frequently shop or they shop in bulk values. The company should
# target these customers so as to make them high value customers. The company can give them freebies and attractive
# discounts so as to make them shop more.

# Promising customers have the potential to become medium value customers. These customers are those who have purchased 
# long back ago or purchased recently but of very less value or they purchase very rarely. Company should target them
# and give them limited period offers so as to make them visit the website. Company can send the customers recommendations
# based on their wish list items and offer discounts to them.

# Around 18% (774) of the customers are low value customers. These customers have purchased some items long back ago and have
# not come back since, or they went with the company's competitors. Company should assign one support executive to speak
# with those customers and try to find out the reason for their low activity. Maybe it was a bad experience with the
# website or with the product. The company should try to find out the reason and devise one strategy according to the 
# feedback received from the customers.
```

```python
# Now we have to decide the optimum number of clusters (k)
# For this we will use the elbow method to find the k value

# Importing necessary libraries
from sklearn.cluster import KMeans

# Calculating the sum of squared distances for each k value
sum_of_sq_dist = {}
for k in range(1,15):
    km = KMeans(n_clusters= k, init= 'k-means++', max_iter= 1000)
    km = km.fit(data_scaled)
    sum_of_sq_dist[k] = km.inertia_
    
# Plot the graph for the sum of square distance values and Number of Clusters
sns.pointplot(x = list(sum_of_sq_dist.keys()), y = list(sum_of_sq_dist.values()))
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Sum of Square Distances')
plt.title('Elbow Method For Optimal k')
plt.show()

# From the below figure we see that after k = 5, sum of squared distances for other clusters are decreasing in a linear
# fashion. There is not much difference in the decrease of the distances after k =5. We can also take k as 4. 
# But in this case i will go with k = 5.
```

```python
# Building the K-Means clustering model with k = 5
KMean_clust = KMeans(n_clusters= 5, init= 'k-means++', max_iter= 1000)
KMean_clust.fit(data_scaled)

# Find the clusters for the observation given in the dataset
RFMScores['Cluster'] = KMean_clust.labels_
RFMScores.head()

# Now every customer has been classified to one of the 5 clusters (0,1,2,3,4).
```
